### Korelacja zmiennych

```python
correlation = math_students.corr()
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(correlation, annot = True)
plt.show()
```

```python
df=por_students.drop('G2',axis=1)
df=por_students.drop('G1',axis=1)
df.corr().nlargest(20,'G3')['G3']
```

Z macierzy relacji widzimy, że najbardziej skorelowane z G3 jest G2 i G1. Nie jest to zaskakujące gdyż są to oceny 'cząstkowe'. Poza tymi oczywistymi korelacjami najbardziej skorelowane z G3 jest czas uczenia się, a także wykszatałcenie rodziców. Największą korelację na minusie ma natomiast ilość oblanych semestrow i ilość sporzywanego alkoholu.

### Badanie mocno skorelowanych danych

```python
sns.pairplot(math_students[[ 'absences', 'G1', 'G2', 'G3']])
plt.show()
```

```python
sns.regplot(data=por_students, y="Medu", x="Fedu", y_jitter=0.4, x_jitter=0.4, scatter_kws={'alpha':0.2})
plt.show()
```

Widzimy, że ludzie żenią się najczęściej z ludźmi o podobnym wykształceniu. Wykształcenie matki i ojca jes mocno ze sobą skorelowane. Być możę będzie trzeba zrezygnować z jednej kolumny.

```python
mapa = por_students.groupby(["Fedu", "Medu"]).agg({'G3':'mean'}).reset_index()
mapa = mapa.pivot(index='Fedu', columns='Medu', values='G3')
mapa = mapa.iloc[::-1]
sns.heatmap(mapa, annot=True, annot_kws={'size': 12} )
```

Na wykresie wyżej widzimy heatmape wykształceń rodziców i średniej wyników z egzaminu. Wyniki zdają się rosnąć wraz z wykszatałceniami rodziców. Wysokie wartości dla słabego wykszatałcenia rodziców może wynikać z małej ilości rekordów w tych kubełkach co widać na poniższym wykresie.

```python
sns.displot( data = por_students, x="Medu", y="Fedu", bins = 5, cbar=True)
```
