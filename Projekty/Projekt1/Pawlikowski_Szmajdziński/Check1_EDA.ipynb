```python
sns.regplot(data=por_students, y="Medu", x="Fedu", y_jitter=0.4, x_jitter=0.4, scatter_kws={'alpha':0.2})
plt.show()
```

Widzimy, Å¼e ludzie Å¼eniÄ… siÄ™ najczÄ™Å›ciej z ludÅºmi o podobnym wyksztaÅ‚ceniu. WyksztaÅ‚cenie matki i ojca jes mocno ze sobÄ… skorelowane. ByÄ‡ moÅ¼Ä™ bÄ™dzie trzeba zrezygnowaÄ‡ z jednej kolumny.

```python
mapa = por_students.groupby(["Fedu", "Medu"]).agg({'G3':'mean'}).reset_index()
mapa = mapa.pivot(index='Fedu', columns='Medu', values='G3')
mapa = mapa.iloc[::-1]
sns.heatmap(mapa, annot=True, annot_kws={'size': 12} )
```

Na wykresie wyÅ¼ej widzimy heatmape wyksztaÅ‚ceÅ„ rodzicÃ³w i Å›redniej wynikÃ³w z egzaminu. Wyniki zdajÄ… siÄ™ rosnÄ…Ä‡ wraz z wykszataÅ‚ceniami rodzicÃ³w. Wysokie wartoÅ›ci dla sÅ‚abego wykszataÅ‚cenia rodzicÃ³w moÅ¼e wynikaÄ‡ z maÅ‚ej iloÅ›ci rekordÃ³w w tych kubeÅ‚kach co widaÄ‡ na poniÅ¼szym wykresie.

```python
sns.displot( data = por_students, x="Medu", y="Fedu", bins = 5, cbar=True)
```
