## Pakiety

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split

import warnings
warnings.filterwarnings('ignore')
np.random.seed = 42
```

## Wczytanie zbioru danych

```python
data = pd.read_csv('australia.csv')
data
```

```python
data.describe()
```

Widać, że mamy pewne niezrównoważenie w danych. Tylko 22% obserwacji wkazuje na wystąpenie deszczu kolejnego dnia.

## Podział na zbiór traningowy i testowy

```python
y = np.array(data['RainTomorrow'])
X = data.drop(['RainTomorrow'],axis=1)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Uczenie 3 klasyfikatorów

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=42, max_iter=1000, solver='sag')
lr.fit(X_train,y_train)
y_hat_lr = lr.predict(X_test)
```

```python
#LogisticRegression?
```

Parametr solver określa, jaki algorytm zostanie użyty w problemie optymalizacji. Zmiana z domyślnego "lbfgs" na "sag" wpłynęła na drobną poprawę otrzymywanych wyników.
Prócz tego podniosłem maksymalną liczbę iteracji do 1000.

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=42, max_depth = 7)
tree.fit(X_train,y_train)
y_hat_tree = tree.predict(X_test)
```

Parametr max_depth określa maksymalną wysokość drzewa. w przeciwnym wypadku podziały są dokonywane, aż liście będą oczyszczone lub przekroczona zostanie liczba min_samples_split.
To dla wysokości 7 uzsykałem najbardziej obiecujące wyniki

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
grad = GradientBoostingClassifier(random_state=42,
                                  learning_rate=0.3)
grad.fit(X_train, y_train)
y_hat_grad = grad.predict(X_test)
```

Paramter learning_rate obniża o stosowną wartość wpływ, jaki posiada każdy nowododany model bazowy (do naszej serii modelów), na wynik predykcji.
Zwiększenie wartości domyślnej 0.1 na 0.3 poprawiło wyniki.

## Metryki

```python
from sklearn.metrics import accuracy_score as acc
from sklearn.metrics import f1_score as f1
from sklearn.metrics import plot_roc_curve as roc_plot
from sklearn.metrics import roc_auc_score as auc
```

```python
print("Accuracy score:", acc(y_test, y_hat_lr))
print("F1 score:", f1(y_test, y_hat_lr))
print("Weighted F1 score:", f1(y_test, y_hat_lr, average='weighted'))
print("Area under the curve:", auc(y_test, y_hat_lr))
roc_plot(lr, X_test, y_test)
plt.show()
```

```python
print("Accuracy score:", acc(y_test, y_hat_tree))
print("F1 score:", f1(y_test, y_hat_tree))
print("Weighted F1 score:", f1(y_test, y_hat_tree, average='weighted'))
print("Area under the curve:", auc(y_test, y_hat_tree))
roc_plot(tree, X_test, y_test)
plt.show()
```

```python
print("Accuracy score:", acc(y_test, y_hat_grad))
print("F1 score:", f1(y_test, y_hat_grad))
print("Weighted F1 score:", f1(y_test, y_hat_grad, average='weighted'))
print("Area under the curve:", auc(y_test, y_hat_grad))
roc_plot(grad, X_test, y_test)
plt.show()
```

```python
tablica_metryk = ["Accuracy", "Accuracy", "Accuracy", "F1", "F1", "F1", "Weighted F1", "Weighted F1", "Weighted F1", "AUC", "AUC", "AUC"]
wartosci = [acc(y_test, y_hat_lr), acc(y_test, y_hat_tree), acc(y_test, y_hat_grad), f1(y_test, y_hat_lr), f1(y_test, y_hat_tree), f1(y_test, y_hat_grad), f1(y_test, y_hat_lr, average='weighted'), f1(y_test, y_hat_tree, average='weighted'), f1(y_test, y_hat_grad, average='weighted'), auc(y_test, y_hat_lr), auc(y_test, y_hat_tree), auc(y_test, y_hat_grad)]
klasyfikator = ["lr", "tree", "grad"]*4

d = {'Metryka': tablica_metryk, 'Wynik': wartosci, 'Klasyfikator': klasyfikator}
df = pd.DataFrame(data=d)
df
```

```python
sns.set_theme(style="whitegrid")

g = sns.catplot(
    data=df, kind="bar",
    x="Metryka", y="Wynik", hue="Klasyfikator",
    palette="dark", alpha=.6, height=9
)
```

## Podsumowanie

Każda z zastosowanych metryk jednoznacznie wskazała najlepszy z powyższych modeli, okazał się nim GradientBoostingClassifier. Regresja logistyczna zapewniła jednak tylko nieco gorsze wyniki. 

Ponieważ nie znamy dokładnego celu predykcji, moim zdaniem rozsądnym jest użycie metryki F1, gdyż zarówno FP, jak i FN są dla nas jednakowo ważne, tak samo kosztowne. Dodatkowym argumentem przemawiającym na opieraniu się własnie o tę metrykę w ewaluowaniu modeli jest pewne, wspomniane na początku, niezbalansowanie danych.
